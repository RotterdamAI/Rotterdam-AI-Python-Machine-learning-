# Data Visualization using Matplotlib and Seaborn

### Step 1: Import Libraries and Load Data

We'll start by importing the necessary libraries and loading the "iris" dataset, which contains information about different species of iris flowers and their characteristics.

```python
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the iris dataset
iris = sns.load_dataset('iris')
```

### Step 2: Basic Data Exploration

Before we proceed with data visualization, let's perform some basic data exploration to get an overview of the dataset.

```python
# Check the first few rows of the dataset
print(iris.head())

# Summary statistics of the dataset
print(iris.describe())

# Count the number of samples for each species
print(iris['species'].value_counts())
```

### Step 3: Scatter Plot with Matplotlib

A scatter plot is a great way to visualize the relationship between two numerical features. Let's create a scatter plot to examine the relationship between "Petal Length" and "Petal Width."

```python
# Scatter plot using Matplotlib
plt.scatter(iris['petal_length'], iris['petal_width'])
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Scatter Plot: Petal Length vs. Petal Width')
plt.show()
```

### Step 4: Histogram with Matplotlib

Histograms help us understand the distribution of a numerical feature. Let's create a histogram to visualize the distribution of "Sepal Width."

```python
# Histogram using Matplotlib
plt.hist(iris['sepal_width'], bins=15, edgecolor='black')
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
plt.title('Histogram: Sepal Width')
plt.show()
```

### Step 5: Box Plot with Seaborn

Box plots are useful for visualizing the distribution of a numerical feature across different categories. Let's create a box plot to compare the "Petal Length" for each species of iris.

```python
# Box plot using Seaborn
sns.boxplot(x='species', y='petal_length', data=iris)
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.title('Box Plot: Petal Length by Species')
plt.show()
```

### Step 6: Violin Plot with Seaborn

A violin plot is similar to a box plot but provides a more detailed view of the distribution. Let's create a violin plot to compare the "Petal Width" for each species of iris.



```python
# Violin plot using Seaborn
sns.violinplot(x='species', y='petal_width', data=iris)
plt.xlabel('Species')
plt.ylabel('Petal Width')
plt.title('Violin Plot: Petal Width by Species')
plt.show()
```

### Step 7: Pair Plot with Seaborn

Pair plots display pairwise relationships between numerical features in the dataset. Let's create a pair plot for all numerical features in the iris dataset, color-coded by species.

```python
# Pair plot using Seaborn
sns.pairplot(iris, hue='species')
plt.suptitle('Pair Plot: Iris Dataset', y=1.02)
plt.show()
```

### Step 8: Heatmap with Seaborn

Heatmaps are useful for visualizing the correlation between numerical features in a dataset. Let's create a correlation heatmap for the iris dataset.

```python
iris['species'].replace(['setosa', 'versicolor', 'virginica'],
                        [0, 1, 2], inplace=True)

iris
```

```python
# Calculate the correlation matrix
correlation_matrix = iris.corr()

# Heatmap using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

### Step 9: Bar Plot with Seaborn

Bar plots are useful for comparing numerical values across different categories. Let's create a bar plot to compare the mean "Sepal Length" for each species of iris.

```python
# Bar plot using Seaborn
sns.barplot(x='species', y='sepal_length', data=iris, ci=None)
plt.xlabel('Species')
plt.ylabel('Mean Sepal Length')
plt.title('Bar Plot: Mean Sepal Length by Species')
plt.show()
```

### Step 10: Count Plot with Seaborn

Count plots display the number of occurrences of each category in a categorical feature. Let's create a count plot to visualize the distribution of different iris species in the dataset.

```python
# Count plot using Seaborn
sns.countplot(x='species', data=iris)
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Count Plot: Distribution of Iris Species')
plt.show()
```

These are just a few examples of the many types of data visualizations you can create using Matplotlib and Seaborn. These libraries offer a wide range of customization options to help you effectively explore and present your data.
